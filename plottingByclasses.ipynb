```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
# only set one to True
density = False
strict = False
best = False
vertical = False
```

```python
file2rows = ""
file4rows = ""
file24rows = ""
```

```python
prefix = ''
if(density == True):
    prefix = 'densityFit/'
elif (strict == True):
    prefix = 'strictFit/'
elif(vertical == True):
    prefix = 'verticalFit/'
elif(best == True):
    prefix = 'bestFit/' #not created yet
    
file4rows = prefix + 'filter3_4rows.csv'
file2rows = prefix + 'filter3_2rows.csv'
file24rows = prefix + 'filter3_2_4_rows.csv'    
    
```

```python
df4 = pd.read_csv(file4rows)
df2 = pd.read_csv(file2rows)
df24 = pd.read_csv(file24rows)
```

```python
def boxplot(x_axis,y_axis,dataframe):
    sns.set_style("whitegrid") 
    sns.boxplot(x = x_axis, y = y_axis, data = dataframe) 
```

```python
#box plot and density figure of 4rows plots by binary classification
# >=10% setting to 1
plt.figure()
boxplot('class10','TotalMSE',df4)
plt.figure()
df4.groupby('class10').TotalMSE.plot(kind='kde')
```
